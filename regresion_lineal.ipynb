```
import numpy as np
import scipy as sc
from matplotlib import pyplot as plt
from sklearn.datasets import load_boston  # Importamos un dataset de testing
```

```
# cargamos la librería en una variable
boston = load_boston()
print(boston.DESCR)
```

```
# estimación usando minimos cuadrados simples
# beta = (X.T*X)^-1*X.T*Y
```

```
X = np.array(boston.data[:,5])
Y = np.array(boston.target)
X.shape

plt.scatter(X, Y, alpha=0.3)

# Añadimos una columna de unos para tener en cuenta los términos independientes
UNOS = np.ones(506)
X = np.array([UNOS, X]).T  # ponemos 506 porque sabemos que son los datos.
# @ para python es la multiplicación matricial.
B = np.linalg.inv(X.T@X) @ X.T @ Y

b1 = B[0] + B[1] * 4
b2 = B[0] + B[1] * 9
plt.plot([4,9],[b1,b2], c = 'red')
plt.show()
```

Básicamente lo que nos dice el modelo es que a más habitaciones mayor es el valor del alquiler, algo muy simple de ver a simple vista.
